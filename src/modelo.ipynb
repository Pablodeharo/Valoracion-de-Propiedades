Diseño del modelo de deep learning.

Utilizaremoes una arquitectura que combine diferentes tipos de capas:
1. Capa de entrada: Para características individuales de la propiedad.
2. Capa de embedding: Para el número de hoja catastral.
3. Capas convolucionales: Para capturar patrones especiales.
4. Capas LSTM: Para tendencias tendenias temporales (Si tenemos datos históricos)
5. Capas densas: Para la predicción final

Anatomia de una grafica de Matplotlib

Figura

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Datos a graficar
x = np.linspace(0,10*np.pi,400)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y);

figura = plt.gcf() # Get current figure
type(figura)
```

```python
plt.plot(x,y);

axes = plt.gca()
type(axes)

```

```python
# crear la figura y los axes
fig, ax = plt.subplots()

ax.plot(x,y)
```
